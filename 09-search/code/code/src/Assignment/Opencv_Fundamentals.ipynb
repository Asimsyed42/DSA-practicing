```python
import cv2
from google.colab.patches import cv2_imshow

# Load the image
image = cv2.imread('modi.jpg')

# Display the image
cv2_imshow(image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import matplotlib.pyplot as plt
import cv2
from google.colab.patches import cv2_imshow

gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap='gray')
plt.title("Grayscale Image")
plt.axis('off')
plt.show()
```

```python
import matplotlib.pyplot as plt
import cv2
from google.colab.patches import cv2_imshow
resized = cv2.resize(image, (200, 200))
plt.imshow(cv2.cvtColor(resized, cv2.COLOR_BGR2RGB))
plt.title("Resized Image (200x200)")
plt.axis('off')
plt.show()
```

```python
import matplotlib.pyplot as plt
import cv2
from google.colab.patches import cv2_imshow

cropped = image[50:200, 100:300]  # crop part of the image
plt.imshow(cv2.cvtColor(cropped, cv2.COLOR_BGR2RGB))
plt.title("Cropped Image")
plt.axis('off')
plt.show()
```

```python
import matplotlib.pyplot as plt
import cv2
from google.colab.patches import cv2_imshow
draw_img = image.copy()
cv2.rectangle(draw_img, (50, 50), (200, 200), (0, 255, 0), 3)  # rectangle
cv2.circle(draw_img, (300, 150), 40, (255, 0, 0), -1)          # circle
cv2.line(draw_img, (0, 0), (300, 300), (0, 0, 255), 2)         # line
plt.imshow(cv2.cvtColor(draw_img, cv2.COLOR_BGR2RGB))
plt.title("Draw Shapes")
plt.axis('off')
plt.show()
```

```python
import matplotlib.pyplot as plt
import cv2
from google.colab.patches import cv2_imshow

text_img = image.copy()
cv2.putText(text_img, "Trump -Modi", (150, 250),
            cv2.FONT_HERSHEY_SIMPLEX, 1.5, (255, 255, 0), 3)
plt.imshow(cv2.cvtColor(text_img, cv2.COLOR_BGR2RGB))
plt.title("Trump - Modi")
plt.axis('off')
plt.show()
```

```python
import matplotlib.pyplot as plt
import cv2
from google.colab.patches import cv2_imshow

blur = cv2.GaussianBlur(image, (9,9), 0)
plt.imshow(cv2.cvtColor(blur, cv2.COLOR_BGR2RGB))
plt.title("Blurred Image")
plt.axis('off')
plt.show()
```

```python
import matplotlib.pyplot as plt
import cv2
from google.colab.patches import cv2_imshow

edges = cv2.Canny(image, 100, 200)
plt.imshow(edges, cmap='gray')
plt.title("Edge Detection")
plt.axis('off')
plt.show()
```
